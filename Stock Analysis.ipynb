```python
# parent folder containing all month folders

base_dir = r"E:\DATA SCIENCE\Data-Driven Stock Analysis Project"
```

```python
# List of month folders

month_folders = [
    "2024-01", "2024-02", "2024-03", "2024-04", "2024-05", 
    "2024-06", "2024-07", "2024-08", "2024-09", "2024-10", "2024-11"
]
```

```python
# Collecting all YAML data from all months

all_data = []

for month in month_folders:
    folder_path = os.path.join(base_dir, month)
    yaml_files = glob.glob(os.path.join(folder_path, "*.yaml"))
    
    for file_path in yaml_files:
        with open(file_path, "r") as f:
            data = yaml.safe_load(f)
            if data:
                all_data.extend(data)
        print(f"ðŸ“„ Loaded {os.path.basename(file_path)} from {month}")

print(f"\nâœ… Total records loaded: {len(all_data)}")
```

```python
# Group by ticker

ticker_groups = {}
for record in all_data:
    ticker = record.get("Ticker")
    if ticker not in ticker_groups:
        ticker_groups[ticker] = []
    ticker_groups[ticker].append(record)
```

```python
# Created output directory

output_dir = os.path.join(base_dir, "All_Ticker_CSVs")
os.makedirs(output_dir, exist_ok=True)
```

```python
# Exported each ticker as one CSV

for ticker, records in ticker_groups.items():
    df = pd.DataFrame(records)
    output_path = os.path.join(output_dir, f"{ticker}.csv")
    df.to_csv(output_path, index=False)
    print(f"ðŸ’¾ Saved: {output_path}")

print("\n All monthly YAMLs processed successfully â€” each ticker saved as a single CSV in 'All_Ticker_CSVs'!")
```
